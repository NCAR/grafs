```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
from glob import glob
from pygrafs.evaluation.scores import *
from sklearn.ensemble import GradientBoostingRegressor
```

```python
data = pd.read_csv("/d2/dgagne/merged_data/int_fcst_grid.20141101.05.csv")
data = data[data['av_dswrf_sfc_f'] >= 0]
data = data[data['av_dswrf_sfc'] < 30000]
```

```python
path = "/d2/dgagne/merged_data/"
def load_all_data(path):
    data_files = sorted(glob(path + "*.csv"))
    all_data_list = []
    for data_file in data_files:
        all_data_list.append(pd.read_csv(data_file))
        all_data_list[-1] = all_data_list[-1][(all_data_list[-1]['av_dswrf_sfc_f'] >= 0) & (all_data_list[-1]['av_dswrf_sfc'] < 30000)]
    all_data = all_data_list[0].append(all_data_list[1:],ignore_index=True)
    return all_data
all_data = load_all_data(path)
all_data['local_hour'] = (all_data['valid_hour'] - 8) % 24
day_data = all_data[(all_data['local_hour'] >= 7) & (all_data['local_hour'] <= 17)]
day_data.reset_index(inplace=True)
print day_data.shape
print day_data.index
```

```python
random_indices = np.random.permutation(day_data.shape[0])
split = day_data.shape[0] * 7 / 10
train_indices = random_indices[:split]
test_indices = random_indices[split:]
input_columns = ['valid_hour','day_of_year','sine_doy','lon','lat','T_f','av_dswrf_sfc_f','cloud_cover_f','dewpt_f']
output_column = ['av_dswrf_sfc']
```

```python
gbr = GradientBoostingRegressor(n_estimators=10,loss='lad',learning_rate=0.3,max_depth=10,subsample=0.3)
gbr.fit(day_data.loc[train_indices,input_columns],day_data.loc[train_indices,output_column].values[:,0])
preds = gbr.predict(day_data.loc[test_indices,input_columns])
```

```python
plt.figure(figsize=(8,4))
plt.subplot(1,2,1)
plt.scatter(preds,all_data.loc[test_indices,output_column])
plt.xlim(0,1000)
plt.ylim(0,1000)
plt.xlabel("GBR Forecast")
plt.ylabel("Observed")
plt.subplot(1,2,2)
plt.scatter(all_data.loc[test_indices,'av_dswrf_sfc_f'],all_data.loc[test_indices,output_column])
plt.xlim(0,1000)
plt.ylim(0,1000)
plt.xlabel("NAM Forecast")
plt.suptitle("ML vs. NWP Forecast Comparison")
plt.savefig("/d2/dgagne/grafs_figures/first_forecast_comp.png",dpi=200,bbox_inches='tight')
```
